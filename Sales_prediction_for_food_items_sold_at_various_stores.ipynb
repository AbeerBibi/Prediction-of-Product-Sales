#Loading Data

```python
import pandas as pd
path="/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/sales_predictions_2023.csv"
df_sales = pd.read_csv(path)
df_sales
df_sales.head()

#1) How many rows and columns
df_sales.info()
```

#Data Cleaning

```python
#2) What are the datatypes of each variable?
col_types=df_sales.dtypes
col_types

#3) Are there duplicates? If so, drop any duplicates.
duplicated_rows= df_sales.duplicated()
df_sales[duplicated_rows]
df_sales.duplicated().sum() # there are no duplicates

#4) Identify missing values.
null_values=df_sales.isna()
df_sales[null_values]
df_sales.isna().sum()

#5) Address the missing values by using a placeholder value.
df_sales['Item_Weight']=df_sales['Item_Weight'].fillna(-1) # Item_Weight is numeric colum
df_sales['Outlet_Size']=df_sales['Outlet_Size'].fillna('MISSING') # Outlet_Size is object colum


#6) Confirm that there are no missing values after addressing them.
df_sales.isna().sum() #to chech that there is no null values anymore

#7) Find and fix any inconsistent categories of data (for example: fix cat, Cat, and cats so that they are consistent).
for col in df_sales:
 print(f"Value Counts for {col}")
 print(df_sales[col].value_counts())
 print('\n') #no inconsistentcy

#8) For any numerical columns, obtain the summary statistics of each (min, max, mean).
df_sales.describe()
```
