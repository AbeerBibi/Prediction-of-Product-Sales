#For Part 3, your task is to create exploratory visuals that might help you understand, explain, or model your data. This must include at least one of each:



#Histograms to view the distributions of numerical features in your dataset.


```python
numeric_columns = df.select_dtypes('number')
numeric_columns
```

```python
#use for loop to iterate for each colum
for column in numeric_columns.columns:
    sns.histplot(data=numeric_columns, x=column, kde=True, bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

##Boxplots to view statistical summaries of numerical features in your dataset.

```python
for column in numeric_columns.columns:
    sns.boxplot(data=numeric_columns, x=column)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

##Countplots to view the frequency of each class of categorial features in your dataset.

```python
#object colums
cat_columns = df.select_dtypes('object').drop(columns=['Item_Identifier'])
cat_columns
```

```python
for column in cat_columns.columns:
    sns.countplot(data=cat_columns, x=column)
    plt.title(f'Count of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45,ha="right" )
    plt.show()
    print("\n")
```

##Heatmap to view the correlation between features.

```python
# Find correlations
corr = df.corr(numeric_only = True)
# Make a heatmap using the correlations from above
sns.heatmap(corr, cmap = 'Reds', annot = True);
```
